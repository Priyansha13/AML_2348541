```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
```

```python
x = 6*np.random.rand(100,1)-3
y=0.5 * x**2 +1.5*x+2+np.random.randn(100,1)
plt.scatter(x,y,color='b')
plt.xlabel('X Points')
plt.ylabel('Y Points')
```

x = np.arange(0,25)

```python
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
reg_Model1= LinearRegression()
```

```python
reg_Model1.fit(x_train, y_train)
```

```python
score1 = r2_score(y_test, reg_Model1.predict(x_test))
print(score1)
```

```python
reg_Model1.coef_
```

```python
reg_Model1.intercept_
```

```python
plt.plot(x_train, reg_Model1.predict(x_train), color='r')
plt.scatter(x_train, y_train)
plt.xlabel('X Values')
plt.ylabel('Y Values')
```

### Polynomial Regression

```python
p= PolynomialFeatures(degree=2, include_bias=True)
x_train_p = p.fit_transform(x_train)
x_test_p = p.transform(x_test)
```

```python
x_train_p
```

```python
x_test_p
```

```python
reg_Model2 = LinearRegression()
reg_Model2.fit(x_train_p, y_train)
y_predict2 = reg_Model2.predict(x_test_p)
score2 = r2_score(y_test, y_predict2)
```

```python
score2
```

```python
print(reg_Model2.coef_)
print(reg_Model2.intercept_)
```

```python
plt.plot(x_train, reg_Model2.predict(x_train_p))
plt.scatter(x_train, y_train)
```

```python
x = 6*np.random.rand(250,1)-3
y=0.5 * x**2 +1.5*x+2+np.random.randn(250,1)
plt.scatter(x,y,color='b')
plt.xlabel('X Points')
plt.ylabel('Y Points')
```

```python
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.2, random_state=2)
p= PolynomialFeatures(degree=2, include_bias=True)
x_train_p = p.fit_transform(x_train)
x_test_p = p.transform(x_test)
```

```python
reg_Model3 = LinearRegression()
reg_Model3.fit(x_train_p, y_train)
y_predict3 = reg_Model3.predict(x_test_p)
score2 = r2_score(y_test, y_predict3)
```

```python
plt.plot(x_train, reg_Model3.predict(x_train_p))
plt.scatter(x_train, y_train)
plt.xlabel('X Values')
plt.ylabel('Y Values')
```
